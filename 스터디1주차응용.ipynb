```python
import pandas as pd
```

```python
# 파일 경로를 찾고, 변수에 저장
file_path = './NetFlix.csv'

# read_csv() 함수로 데이터프레임 변환
df = pd.read_csv(file_path)
df.head()
```

```python
# Q1 : release_year = 2020 인 영화 출력


year = df[df['release_year'] == 2020]
year.head()
```

```python
# Q2 : release_year이 2020년이면서, duration이 1이 아닌 데이터

dur = year[year['duration']!=1]
dur.head()
```

```python
# Q3 : Q2에서 duration 순으로 내림차순 후 director=NaN 빼기

asc = dur.sort_values(by='duration',ascending=False)
asc_na = asc.dropna(subset=['director'], axis=0)
asc_na
```

```python
# Q4 : type=TVshow, release_year 기준으로 정렬, 결측치가 있는 것만 출력

type_tv = df[df['type']== 'TV Show']
type_tv2 = type_tv.sort_values(by='release_year',ascending=True)
type_tv3 = type_tv2[type_tv2.isna().any(axis=1)]
type_tv3.info()
```

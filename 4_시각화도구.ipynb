```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Matplotlib - 기본 그래프 도구

```python
# matplotlib 한글 폰트 오류 문제 해결
from matplotlib import font_manager, rc
font_path = "./malgun.ttf"
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font',family=font_name)
```

```python
# 선 그래프

df = pd.read_excel('시도별 전출입 인구수.xlsx', header=0)

df = df.fillna(method='ffill') # 누락값(NaN)을 앞 데이터로 채움(엑셀 양식 병합 부분)

mask = (df['전출지별'] == '서울특별시') & (df['전입지별'] != '서울특별시')
df_seoul = df[mask]
df_seoul = df_seoul.drop(['전출지별'],axis=1)
df_seoul.rename({'전입지별':'전입지'},axis=1,inplace=True)
df_seoul.set_index('전입지',inplace=True)

sr_one = df_seoul.loc['경기도'] # 서울에서 경기도로 이동한 인구 데이터 값만 선택
plt.plot(sr_one.index, sr_one.values) # or plt.plot(sr_one)

plt.title('서울 -> 경기 인구 이동') # 차트 제목 추가

# 축 이름 추가
plt.xlabel('기간')
plt.ylabel('이동 인구수')
```

```python
# 그래프 꾸미기

sr_one = df_seoul.loc['경기도']

plt.figure(figsize=(14,5)) # 그림 사이즈 지정(가로 14, 세로 5)
plt.xticks(rotation='vertical') # x축 눈금 라벨 회전
plt.plot(sr_one)

plt.title('서울 -> 경기 인구 이동')
plt.xlabel('기간')
plt.ylabel('이동 인구수')

plt.legend(labels=['서울 -> 경기'],loc='best') # 범례 표시
plt.show()
```

```python
# 스타일 서식 지정

sr_one = df_seoul.loc['경기도']

plt.style.use('ggplot') # 스타일 서식 지정

plt.figure(figsize=(14,5)) # 그림 사이즈 지정(가로 14, 세로 5)
plt.xticks(size=10, rotation='vertical') # x축 눈금 라벨 회전
plt.plot(sr_one,marker='o',markersize=10)

plt.title('서울 -> 경기 인구 이동',size=30)
plt.xlabel('기간',size=20)
plt.ylabel('이동 인구수',size=20)

plt.legend(labels=['서울 -> 경기'],loc='best',fontsize=15) # 범례 표시
plt.show()
```
